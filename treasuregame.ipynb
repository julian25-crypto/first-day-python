```python
def start_game():
  print("Welcome to Treasure game")
  print("You have hit a crossroad")
  print("river or forest")

  choice = input("Where do you want to go?").lower()
  if choice == "river":
    river()
  elif choice == "forest":
    forest()
  else:
    print("Invalid input")


def forest():
  print("Welcome to forest")
  print("You have hit a crossroad")
  print("cave or light")

  choice = input("Where do you want to go?").lower()
  if choice == "cave":
    print("You found treasure in cave")
  elif choice == "light":
    print("You got scared by light. No treasure found. Start again")
  else:
    print("Invalid input")


def river():
  print("Welcome to the river")
  print("You have hit a crossroad")
  print("stream or fisherman")

  choice = input("Where do you want to go?").lower()
  if choice == "stream":
    print("You found treasure in water")
  elif choice == "fisherman":
    print("You got scared by him. Try again")
  else:
    print("Invalid input")


start_game()


```
