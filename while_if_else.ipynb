Movie booking system


```python
print("Welcome to movie booking system")

available_seats = 5
password = "2007" #for manager who is booking seats
```

```python
print("What do you want to do")
print("1. Book tickets")
print("2. Check available seats")
print("3. Admin add seats")
print("4. Exit")
```

```python
while True:
  choice = int(input("What is your choice "))

  if choice == 1:
    print("Tickets booked")
  elif choice == 2:
    print("Available seats" , available_seats)
  elif choice == 3:
    admin_pass = input("Enter your password to add seats ")
    if admin_pass == password:
      new_seats = int(input("How many seats you want to add "))
      available_seats = available_seats + new_seats
      print("Seats added successfully")
      print("Current available seats " , available_seats)
    else:
      print("Incoorect password entered")
  elif choice == 4:
    print("Exit")
    break
  else :
    print("Not a valid choice")
```
