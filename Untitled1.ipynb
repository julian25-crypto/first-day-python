GUESS GAME

```python
import random # choose a random number automatically
print("Welcome to the guessing world")


secret_number = random.randint(1,20)
# terminate the game , if user guesses correctly or 5 attempts over
for i in range(5):
  uno =   int(input("Guess the number "))
  if uno == secret_number:
    print("CORRECT GUESS")
    break
  elif uno > secret_number:
    print("TOO HIGH")
  elif uno < secret_number:
    print("TOO LOW")
```
