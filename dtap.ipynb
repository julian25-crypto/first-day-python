dating app

```python
def create_profile(name , age , interest , city):
  profile = {
    "name" : name,
    "age" : age,
    "interest" : interest,
    "city" : city
  }
  return profile

create_profile("Ashish" , 24 , "travel" , "pune")

def calculate_match_score(profile1 , profile2):
  score = 0

  #same interest
  if profile1["interest"] == profile2["interest"]:
    score = score + 50

  #same city
  if profile1["city"] == profile2["city"]:
    score = score + 30

  age_diff = abs(profile1["age"] - profile2["age"])
  if age_diff <= 2 :
    score += 20

  return score


def is_match(score):
  if score >= 60:
    return True
  else :
    return False


def display_results(profile1 , profile2 , score):
  print("Welcome to dating app")
  print("=" *20)
  print("Person 1 " , profile1["name"])
  print("Person 2 " , profile2["name"])
  print("Compatibility score" , score)

  if is_match(score):
    print("It's a Match. Woohoo")
  else:
    print("Oops! You are still single, keep moving.")


user1 = create_profile("Ashish" , 25 , "travel" , "delhi")
user2 = create_profile("Ashika" , 23 , "travel" , "pune")

match_score = calculate_match_score(user1 , user2)

display_results(user1 , user2 , match_score)



```
